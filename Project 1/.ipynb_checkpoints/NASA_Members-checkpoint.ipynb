```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
nasa = pd.read_csv('../Resources/NASA_GEOSS2S_MEMBER_LDC.csv')
nasa['MON'] = pd.to_datetime(nasa['MON'], format = '%H%M %d %b %Y')
nasa['START'] = nasa['MON'].dt.month
nasa['TARGET'] = ((nasa['START'] + nasa['L']).astype(int))%12
nasa['TARGET'] = nasa['TARGET'].apply(lambda x: 12 if x == 0 else x)

# Calculate the mean grouped by 'L' and 'START'
nasa_mean = nasa.groupby(['L', 'START', 'M'])['LDC_SST_NASA_GEOSS2S'].mean().reset_index()
nasa_mean.rename(columns={'LDC_SST_NASA_GEOSS2S': 'MEAN_SST_NASA_GEOSS2S'}, inplace=True)

# Calculate the variance grouped by 'L' and 'START'
nasa_var = nasa.groupby(['L', 'START', 'M'])['LDC_SST_NASA_GEOSS2S'].var().reset_index()
nasa_var.rename(columns={'LDC_SST_NASA_GEOSS2S': 'VAR_SST_NASA_GEOSS2S'}, inplace=True)

# Merge the mean and variance DataFrames
nasa_stat = pd.merge(nasa_mean, nasa_var, on=['M','L','START'])
month_to_season = {
    1: 'DJF', 2: 'JFM', 3: 'FMA', 4: 'MAM', 5: 'AMJ', 6: 'MJJ',
    7: 'JJA', 8: 'JAS', 9: 'ASO', 10: 'SON', 11: 'OND', 12: 'NDJ'
}
nasa_stat['TARGET'] = ((nasa_stat['START'] + nasa_stat['L']).astype(int))%12
nasa_stat['TARGET'] = nasa_stat['TARGET'].apply(lambda x: 12 if x == 0 else x)
nasa_stat['SZN'] = nasa_stat['TARGET'].map(month_to_season)

new_order = ['M','L','START','TARGET','SZN','MEAN_SST_NASA_GEOSS2S','VAR_SST_NASA_GEOSS2S']
nasa_stat = nasa_stat[new_order]

nasa_stat.to_csv('../Resources/nasa_member_stat.csv', index=False)
nasa_stat
```

```python
#code that obtains the observed data and creates a doubled df ready for plotting
obs_nino3_4 = pd.read_csv('../Resources/obs_nino3_4.csv')
bottom_row = obs_nino3_4.iloc[1:]
rest_of_df = obs_nino3_4.iloc[:1]
obs_nino3_4 = pd.concat([bottom_row, rest_of_df]).reset_index(drop=True)

nasa_mean = pd.read_csv('../Resources/nasa_stat.csv')

lead_times = [1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5, 9.5, 10.5]
nasa_mean_dfs = {}

# Loop through lead times
for i in lead_times:
    filtered_nasa = nasa_mean[nasa_mean['L'] == i].copy()
    lead = int(i)
    j=1;
    
    while j < lead: 
        bottom_row = filtered_nasa.iloc[-1:]
        # Extract the DataFrame except the bottom row
        rest_of_df = filtered_nasa.iloc[:-1]
        # Concatenate the bottom row to the top of the rest of the DataFrame
        filtered_nasa = pd.concat([bottom_row, rest_of_df]).reset_index(drop=True)
        j += 1
    
    nasa_mean_dfs[f'lead{lead}'] = filtered_nasa
nasa_mean_dfs['lead1']
```

```python
selected_members = [1, 2, 3, 4]  # Members you want to include
selected_seasons = ['JFM','FMA','MAM', 'AMJ', 'MJJ', 'JJA','JAS','ASO','SON','OND','NDJ','DJF']  # Seasons you want to include
#selected_seasons = ['JFM']
lead_time = 7.5
#lead_time_min = 1.5
#lead_time_max = 7.5

# Filter the DataFrame based on the criteria
nasa_stat_filtered = nasa_stat.loc[
    (nasa_stat['M'].isin(selected_members)) &
    (nasa_stat['SZN'].isin(selected_seasons)) &
    (nasa_stat['L'] == lead_time)
]
nasa_mean_filtered = nasa_mean.loc[
    (nasa_mean['L'] == lead_time)&
    (nasa_mean['SZN'].isin(selected_seasons))
]

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.figure(figsize=(12, 8))

# Create the box plot with the filtered data
sns.boxplot(data=nasa_stat_filtered, x='SZN', y='MEAN_SST_NASA_GEOSS2S', color = 'lightsalmon', order=selected_seasons,
            #notch=True,
            whiskerprops={'color': 'black', 'linewidth': 1}, 
            capprops={'color': 'black', 'linewidth': 1},
            boxprops={'edgecolor': 'black', 'linewidth': 1})
sns.stripplot(data=nasa_mean_filtered, x='SZN', y='MEAN_SST_NASA_GEOSS2S', color='red', dodge=True, alpha=1, order=selected_seasons,size=6)
sns.stripplot(data=obs_nino3_4, x='SZN', y='NINO3.4_L3', color='black', dodge=True, alpha=1, order=selected_seasons, size=6)

plt.title(f'Distribution of NINO3.4 SST Prediction by Members, NASA_GEOSS2S, Lead Time [{lead_time-1}]', size = 22)
plt.xlabel('Season', size = 20)
plt.ylabel('SST (Â°C)', size = 20)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.ylim(ymin=24., ymax=29)
#plt.legend()
plt.show()
```
