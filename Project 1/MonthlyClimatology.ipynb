```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
onset = pd.read_csv('../Resources/EnsoOnset.csv')
```

```python
onset
```

```python
anomalies = pd.read_csv('../Resources/Anomalies.csv')
anomalies
```

```python
anomalies_pivot = anomalies.pivot(index='YR', columns='SEAS', values='ANOM').reset_index()

# Rename columns to match enso_onset_df format
anomalies_pivot.columns.name = None
anomalies_pivot.columns = ['Year'] + list(anomalies_pivot.columns[1:])

# Replace the values in enso_onset_df with those from anomalies_pivot
onset_updated = onset.copy()
onset_updated.update(anomalies_pivot)

onset_updated
```

```python
onset_updated_2 = onset_updated.iloc[ :, 1:13]

onset_updated_2.index = onset_updated['Year']

plot = onset_updated_2.plot(kind = 'bar', figsize = (40,10), width=1)
plt.xticks(rotation=45)
plt.tight_layout()
plot.set_title('ENSO from 1980 to 2023')
plt.axhline(y=0.5, color = 'red')
plt.axhline(y=-0.5, color = 'blue')
```
