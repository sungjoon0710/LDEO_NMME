```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce
gfdl_old_stat = pd.read_csv('../Resources/gfdl_stat.csv')
gfdl_member_stat = pd.read_csv('../Resources/gfdl_member_stat.csv')

nasa_old_stat = pd.read_csv('../Resources/nasa_stat.csv')
nasa_member_stat = pd.read_csv('../Resources/nasa_member_stat.csv')

canada_old_stat = pd.read_csv('../Resources/canada_stat.csv')
canada_member_stat = pd.read_csv('../Resources/canada_member_stat.csv')

canesm_old_stat = pd.read_csv('../Resources/canesm_stat.csv')
canesm_member_stat = pd.read_csv('../Resources/canesm_member_stat.csv')

nemo_old_stat = pd.read_csv('../Resources/nemo_stat.csv')
nemo_member_stat = pd.read_csv('../Resources/nemo_member_stat.csv')

cola_old_stat = pd.read_csv('../Resources/cola_stat.csv')
cola_member_stat = pd.read_csv('../Resources/cola_member_stat.csv')

ncep_old_stat = pd.read_csv('../Resources/ncep_stat.csv')
ncep_member_stat = pd.read_csv('../Resources/ncep_member_stat.csv')
```

```python

# Calculate the mean grouped by 'L' and 'START' - Nino3.4
gfdl_mean = gfdl_member_stat.groupby(['L', 'START'])['MEAN_SST_GFDL_SPEAR'].mean().reset_index()

# Calculate the variance grouped by 'L' and 'START' - Nino3.4
gfdl_var = gfdl_member_stat.groupby(['L', 'START'])['VAR_SST_GFDL_SPEAR'].mean().reset_index()


# Merge the mean and variance DataFrames
dfs = [gfdl_mean, gfdl_var]
gfdl_new_stat = reduce(lambda left, right: pd.merge(left, right, on=['L', 'START']), dfs)
gfdl_new_stat
```

```python
gfdl_old_stat['VAR_SST_GFDL_SPEAR'] = gfdl_new_stat['VAR_SST_GFDL_SPEAR']
gfdl_old_stat
```

```python
gfdl_old_stat.to_csv('../Resources/gfdl_stat.csv', index=False)
```
