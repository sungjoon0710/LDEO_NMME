```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

obs = pd.read_csv('../Resources/ObsSST.csv')
obs = obs.drop([obs.index[0],obs.index[-1]])
obs
```

```python
#Calculating rolling averages of monthly data to create seasonal data set across all years
obs['NINO3.4_L3']=obs['NINO3.4'].rolling(window=3, center=True).mean()
obs['NINO1+2_L3']=obs['NINO1+2'].rolling(window=3, center=True).mean()
obs['NINO3_L3']=obs['NINO3'].rolling(window=3, center=True).mean()
obs['NINO4_L3']=obs['NINO4'].rolling(window=3, center=True).mean()
obs = obs.drop([obs.index[0],obs.index[-1]])
SZN = ['DJF','JFM', 'FMA', 'MAM', 'AMJ', 'MJJ', 'JJA', 'JAS', 'ASO', 'SON', 'OND','NDJ']
obs['SZN'] = SZN * 30
obs
```

```python
#Grouping by month (or season) to Calculate the Mean Seasonal Indices across the 30 year period
monthly_obs = obs.groupby('MON')['NINO1+2_L3'].mean().reset_index()

monthly_nino3 = obs.groupby('MON')['NINO3_L3'].mean().reset_index()
monthly_obs['NINO3_L3'] = monthly_nino3['NINO3_L3']

monthly_nino3_4 = obs.groupby('MON')['NINO3.4_L3'].mean().reset_index()
monthly_obs['NINO3.4_L3'] = monthly_nino3_4['NINO3.4_L3']

monthly_nino4 = obs.groupby('MON')['NINO4_L3'].mean().reset_index()
monthly_obs['NINO4_L3'] = monthly_nino4['NINO4_L3']

SZN = ['DJF','JFM', 'FMA', 'MAM', 'AMJ', 'MJJ', 'JJA', 'JAS', 'ASO', 'SON', 'OND','NDJ']
monthly_obs['SZN'] = SZN
monthly_obs
```

```python
#calculating variance in data
var_nino1_2 = obs.groupby('MON')['NINO1+2_L3'].var().reset_index()
monthly_obs['VARIANCE1+2_L3'] = var_nino1_2['NINO1+2_L3']

var_nino3 = obs.groupby('MON')['NINO3_L3'].var().reset_index()
monthly_obs['VARIANCE3_L3'] = var_nino3['NINO3_L3']

var_nino3_4 = obs.groupby('MON')['NINO3.4_L3'].var().reset_index()
monthly_obs['VARIANCE3.4_L3'] = var_nino3_4['NINO3.4_L3']

var_nino4 = obs.groupby('MON')['NINO4_L3'].var().reset_index()
monthly_obs['VARIANCE4_L3'] = var_nino4['NINO4_L3']

monthly_obs
```

```python
#shifting the top and bottom rows so that we can plot from JFM 
bottom_row = monthly_obs.iloc[1:]
rest_of_df = monthly_obs.iloc[:1]
        
monthly_obs = pd.concat([bottom_row, rest_of_df]).reset_index(drop=True)
monthly_obs
```

```python
monthly_obs.to_csv('../Resources/monthly_obs.csv', index=False)
```

```python
def obs_sst_visualizer(region):
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['Arial']
    plt.figure(figsize=(13, 7))
    
    plt.plot(monthly_obs['SZN'], monthly_obs[f'NINO{region}_L3'], marker='o',markersize=8, linewidth=4, linestyle='-', color='black', label = 'OBS')
    
    # Customize the plot
    plt.title(f'Seasonal NINO{region} SST: Observed Mean 1991-2020', size = 24)
    plt.xlabel('Seasons', size = 20)
    plt.ylabel('SST (Â°C)', size = 20)
    plt.xticks(size = 20)
    plt.yticks(size = 20)
    plt.grid(True)
    plt.legend(frameon=False)
    plt.ylim(ymin=20, ymax = 28)
    plt.savefig(f'../Plots/Obs/observedSST_NINO{region}_region.png', format='png')
    # Show the plot
    plt.show()

regions = ['1+2','3','3.4','4']
for i in regions:
    obs_sst_visualizer(i)
```

```python
import seaborn as sns
def all_obs_visualizer():
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['Arial']
    plt.figure(figsize=(13, 7))

    
    plt.plot(monthly_obs['SZN'], monthly_obs[f'NINO3.4_L3'], marker='o',markersize=6, linewidth=2, linestyle='-', color='black', label = 'Mean', zorder=2)
    sns.boxplot(x='SZN', y='NINO3.4_L3', data=obs, color = 'lightsalmon', zorder=1)
    
    # Customize the plot
    plt.title(f'Distribution of Observed Seasonal SST: ENSO ALL 1991-2020', size = 24)
    plt.xlabel('Seasons', size = 20)
    plt.ylabel('SST (Â°C)', size = 20)
    plt.xticks(size = 20)
    plt.yticks(size = 20)
    plt.grid(True)
    plt.legend(frameon=False)
    plt.ylim(ymin=24, ymax = 30)
    plt.savefig(f'../Plots/Obs/all_distribution_and_mean.png', format='png')
    # Show the plot
    plt.show()

    plt.figure(figsize=(12, 6))
all_obs_visualizer()
```

```python
#JFM time series, per year -> run the numbers through -> show time series 1991 -> 2020 JFM -> put through MKtest -> trend value fore ach season. 
#analyze for each season, store it in a column. pass this data set into MKtest. 
obs_focus = obs[['YR', 'MON', 'SZN', 'NINO3.4_L3']]
szn = ['JFM','FMA', 'MAM', 'AMJ','MJJ','JJA','JAS','ASO','SON','OND','NDJ','DJF']

obs_dfs = {}

# Loop through lead times
for i in szn:
    filtered_gfdl = obs_focus[obs_focus['SZN'] == i].copy()
    obs_dfs[f'{i}'] = filtered_gfdl

obs_dfs['AMJ']
```
